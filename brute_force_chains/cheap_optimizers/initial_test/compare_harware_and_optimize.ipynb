```python
import pandas as pd
```

```python
hardware_df = pd.read_csv("/Users/andrewweiland/UCCS_REU/quantum-optimizer-orchestration/brute_force_chains/cheap_optimizers/initial_test/hardware_aware/hardware_aware_test_results.csv")
optimize_df = pd.read_csv("/Users/andrewweiland/UCCS_REU/quantum-optimizer-orchestration/brute_force_chains/cheap_optimizers/initial_test/optimize_only/optimize_only_test_results.csv")
```

```python
merged = pd.merge(hardware_df, optimize_df, on="opt_chain", suffixes=("_hardware", "_optimize"))
```

```python
merged
```

```python
merged["two_qubit_gate_difference"] =  merged["final_two_qubit_gate_count_hardware"] - merged["final_two_qubit_gate_count_optimize"]
merged["depth_difference"] = merged["final_depth_hardware"] - merged["final_depth_optimize"]
merged
```

```python
#Optimize only is on average 26 fewer two qubit gates than hardware aware.
merged["two_qubit_gate_difference"].describe()
```

```python
# Depth is less drastic difference, but still noticable
merged["depth_difference"].describe()
```
