```python
from pytket.qasm import circuit_from_qasm, circuit_to_qasm_str
from pytket.passes import AutoRebase, DecomposeBoxes, FullPeepholeOptimise, RemoveRedundancies, SequencePass
from pytket.circuit import OpType

circuit_path = "/Users/andrewweiland/UCCS_REU/quantum-optimizer-orchestration/benchmarks/ai_transpile/qasm/efficient_su2_8.qasm"

circuit = circuit_from_qasm(circuit_path)
basis_gates = {OpType.Rz, OpType.SX, OpType.X, OpType.CX}

seq_pass = SequencePass([
    DecomposeBoxes(),
    FullPeepholeOptimise(),
    RemoveRedundancies(),
    AutoRebase(basis_gates)
])

seq_pass.apply(circuit)
output_qasm = circuit_to_qasm_str(circuit)


```

```python
from qiskit import QuantumCircuit

optimized_circuit = QuantumCircuit.from_qasm_str(output_qasm)

print("Depth:", optimized_circuit.depth())
print("2Q gates:", optimized_circuit.num_nonlocal_gates())
print("Total gates:", optimized_circuit.size())
```
