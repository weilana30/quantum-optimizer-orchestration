```python
#Gets 2 random local circuits of each type for benchmark analysis.
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../../circuits.csv")
df = df[df["source"] == "local"]
df
```

```python
sample_df = (
    df.groupby("category")
      .apply(lambda x: x.sample(n=min(2, len(x)), random_state=42))
      .reset_index(level=0)
)
sample_df.to_csv("sample_circuits.csv", index=False)
sample_df
```

```python
import shutil

for _, row in sample_df.iterrows():
    local_path = row["qasm_path"]
    path = "../../../quantum-circuit-chaining/" + local_path
    shutil.copy(
        path, 
        "sample_circuits/" + row["category"] + "__" + row["name"] + ".qasm"
    )
        
    
    
```
