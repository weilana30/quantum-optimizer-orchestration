```python
import pandas as pd
```

```python
master_df = pd.read_csv("master.csv")
master_df
```

```python
# Usable Data
# Includes data that only used wisq_rules, wisq_bqskit, and tket
# These methods did not use hardware-awaredness

usable_df = master_df[~master_df["chain_name"].str.contains("qiskit", na=False)].reset_index(drop=True)
usable_df
```

```python
usable_one_opt = usable_df[usable_df["opt_level"] == 1].reset_index(drop=True)
usable_one_opt
```

```python
duplicates = usable_df[usable_df.duplicated(subset=["opt_chain"], keep=False)]
duplicates
```

```python
non_duplicates = usable_df[~usable_df.duplicated(subset=["opt_chain"], keep=False)].reset_index(drop=True)
non_duplicates
```

```python
conflicting = usable_df.groupby("opt_chain").filter(
    lambda g: (
        g["final_depth"].nunique() > 1
        or g["final_two_qubit_gates"].nunique() > 1
        or g["final_total_gates"].nunique() > 1
    )
)
conflicting
```

```python
clean_df = usable_df.groupby("opt_chain").filter(
    lambda g: (
        g["final_depth"].nunique() == 1
        and g["final_two_qubit_gates"].nunique() == 1
        and g["final_total_gates"].nunique() == 1
    )
)
clean_df
```

```python
# All chains that have already been run that were not duplicated
# And only contain wisq and tket optimization

unique_df = clean_df.drop_duplicates(
    subset=["opt_chain"],
    keep="first"
).reset_index(drop=True)
unique_df
```

```python
unique_df.to_csv("usable_optimizer_only.csv", index=False)
```

```python
hardware_df = master_df[master_df["chain_name"].str.contains("qiskit", na=False)].reset_index(drop=True)
hardware_df
```

```python
clean_hardware = hardware_df.groupby("opt_chain").filter(
    lambda g: (
        g["final_depth"].nunique() == 1
        and g["final_two_qubit_gates"].nunique() == 1
        and g["final_total_gates"].nunique() == 1
    )
)
clean_hardware
```

```python
# All chains that have already been run that were not duplicated
# which use hardware-aware qiskit optimizations

unique_hardware = clean_hardware.drop_duplicates(
    subset=["opt_chain"],
    keep="first"
).reset_index(drop=True)
unique_hardware
```

```python
unique_hardware.to_csv("usable_hardware_aware.csv", index=False)
```

```python
circuit_df = pd.read_csv("circuits.csv")
circuit_df
local = circuit_df[circuit_df["source"] == "local"]
local
```

```python
optimizers = ["qiskit_ai", "qiskit_standard", "tket", "wisq_rules", "wisq_bqskit"]
unique_paths = local["qasm_path"].unique()
len(unique_paths)
```

```python
# Creates dataframe of all optimizer only combinations that have
# not been run yet

from itertools import product
import numpy as np
import json
required_df = pd.DataFrame(columns=[
    "parent_circuit_id",
    "parent_circuit",
    "original_circuit_id",
    "original_circuit",
    "original_category",
    "original_circuit_path",
    "chain_name",
    "opt_level",
    "opt_1",
    "opt_2",
    "opt_3",
    "initial_num_qubits",
    "initial_depth",
    "initial_two_qubit_gates",
    "initial_total_gates",
    "final_num_qubits",
    "final_depth",
    "final_two_qubit_gates",
    "final_total_gates",
    "total_runtime_seconds",
    "opt_chain"
])


for circuit_path in unique_paths:
    for chain_length in range(1, 4):
        for chain in product(optimizers, repeat=chain_length):
                parent_circuit_id = -1
                original_circuit_id = -1
                parent_circuit = ""
                chain_name = ""
                original_circuit = ""
                opt_level = -1
                opt_1 = "N/A"
                opt_2 = "N/A"
                opt_3 = "N/A"
                original_circuit_path = circuit_path
                initial_num_qubits = -1
                initial_depth = -1
                initial_two_qubit_gates = 0
                initial_total_gates = 0
                final_num_qubits = 0
                final_depth = 0
                final_two_qubit_gates = 0
                final_total_gates = 0
                total_runtime = 0.0

                original_circuit = circuit_path.split("/")[-1].replace(".qasm", "")

                opt_chain = original_circuit


                for opt in chain:
                     opt_chain += f"__{opt}"
                
                if (unique_df["opt_chain"] == opt_chain).any():
                     continue
                
                original_circuit_id = local[local["qasm_path"] == circuit_path]["id"].values[0]
                parent_circuit = original_circuit
                parent_circuit_id = original_circuit_id
                opt_level = len(chain)
                if chain_length > 0:
                    opt_1 = chain[0]
                if chain_length > 1: 
                    opt_2 = chain[1]
                if chain_length > 2:
                    opt_3 = chain[2]
                initial_num_qubits = local[local["qasm_path"] == circuit_path]["num_qubits"].values[0]
                initial_depth = local[local["qasm_path"] == circuit_path]["initial_depth"].values[0]
                initial_total_gates = local[local["qasm_path"] == circuit_path]["initial_total_gates"].values[0]
                initial_two_qubit_gates = local[local["qasm_path"] == circuit_path]["initial_two_qubit_gates"].values[0]
                original_category = local[local["qasm_path"] == circuit_path]["category"].values[0]


                required_df.loc[len(required_df)] = [
                    parent_circuit_id,
                    parent_circuit,
                    original_circuit_id,
                    original_circuit,
                    original_category,
                    original_circuit_path,
                    chain_name,
                    opt_level,
                    opt_1,
                    opt_2,
                    opt_3,
                    initial_num_qubits,
                    initial_depth,
                    initial_two_qubit_gates,
                    initial_total_gates,
                    final_num_qubits,
                    final_depth,
                    final_two_qubit_gates,
                    final_total_gates,
                    total_runtime,
                    opt_chain
                ]
                     


```

```python
required_df.to_csv("required_optimizer_only.csv", index=False)
```

```python
# Finds current coverage of each original circuit
# Identifies number of existing and missing chains

coverage_df = pd.DataFrame(columns=[
    "original_circuit",
    "possible_chains",
    "existing_chains",
    "missing_chains"])

for circuit_path in unique_paths:
    original_circuit = circuit_path.split("/")[-1].replace(".qasm", "")
    possible_chains = 155
    existing_chains = unique_df[unique_df["original_circuit"] == original_circuit].shape[0]
    missing_chains = possible_chains - existing_chains

    coverage_df.loc[len(coverage_df)] = [
        original_circuit,
        possible_chains,
        existing_chains,
        missing_chains
    ]
coverage_df.head()
```
